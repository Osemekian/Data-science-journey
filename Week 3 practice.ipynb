```python
temp1 = 77
celsius1 = (temp1 - 32) * 5 / 9
print(celsius1)

temp2 = 95
celsius2 = (temp2 - 32) * 5 / 9
print(celsius2)

temp3 = 50
celsius3 = (temp3 - 32) * 5 / 9
print(celsius3)
```

USING FUNCTION TO WRITE A CODE AND REUSE IT

```python
def hct_to_haemoglobin(hct):
  return (hct / 3)

print(hct_to_haemoglobin(55))
print(hct_to_haemoglobin(45))
print(hct_to_haemoglobin(33))
```

```python
def fahrenheit_to_celsius(fahrenheit):
  return (fahrenheit - 32) * 5 / 9

print(fahrenheit_to_celsius(77))
print(fahrenheit_to_celsius(95))
print(fahrenheit_to_celsius(50))
```

From a function's perspective:

A parameter is the variable listed inside the parentheses in the function definition.

An argument is the actual value that is sent to the function when it is called.

for example:

```python
def my_function(name): # name is a parameter
  print("Hello", name)

my_function("Emil") # "Emil" is an argument
```

TWO ARGUMENTS

```python
def patient_name(fname, lname):
  print(fname + " "+ lname)

patient_name("Emil", "Rugby")
```

```python
patient_name("Precious", "Simeons")
```

KEYWORD ARGUMENTS


```python
def record(condition, name):
  print("Patient = ",name + ", Diagnosis = ", condition)

record(condition = "anaemia", name = "Emil Rugby")
```

```python
record(condition = "Diabetes", name = "Ruth Simeons")
```

```python
record( "anaemia", "Emil Rugby")
```

```python
def record(condition, age, name):
  print("Patient = ",name, age,
        ",Diagnosis = ", condition)

record(condition = "diabetes", name = "Emil Rugby,", age = "68 years")
```

```python
record(condition = "anaemia", name = "Joy Stevens,", age = "18 years")
```

```python
def lab_tests(tests):
    for test in tests:
        print(test)

ordered_tests = ["Complete Blood Count (CBC)", "Lipid Panel", "Blood Glucose Test"]

lab_tests(ordered_tests)
```

```python
def lab_sample(sample):
    print("Patient ID:", sample["patient_id"])
    print("Age:", sample["age"])
    print("Test Ordered:", sample["test_type"])

sample1 = {"patient_id": "PT9402","age":23, "test_type": "Urinalysis, PCV, Grouping and crossmatch"}

lab_sample(sample1)
```

```python
sample2 = {"patient_id": "PT5498","age":67, "test_type": "LFT, EUCR, FLP, FBC/MP"}

lab_sample(sample2)
```
