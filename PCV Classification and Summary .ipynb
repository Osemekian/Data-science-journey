Normal
Polycythaemic

sex

classification


Count how many patients are:
anaemic
normal
polycythaemic


Also count by sex:
number of anaemic males
number of anaemic females
number of normal males
number of normal females
number of polycythaemic males
number of polycythaemic females

Print a final summary report


PCV reference ranges:
Male
Anaemic: PCV < 39%
Normal: PCV 40% – 54%
Polycythaemic: PCV > 54%

Female
Anaemic: PCV < 34%
Normal: PCV 35% – 48%
Polycythaemic: PCV > 49%


```python
def classify_pcv(sex, pcv):
    if sex == "Male":
        if pcv < 40:
            return "Anaemic"
        elif 40 <= pcv <= 54:
            return "Normal"
        else:
            return "Polycythaemic"

    elif sex == "Female":
        if pcv < 36:
            return "Anaemic"
        elif 36 <= pcv <= 48:
            return "Normal"
        else:
            return "Polycythaemic"


patient_ID = [
    "Patient_1", "Patient_2", "Patient_3", "Patient_4", "Patient_5",
    "Patient_6", "Patient_7", "Patient_8", "Patient_9", "Patient_10",
    "Patient_11", "Patient_12", "Patient_13", "Patient_14", "Patient_15",
    "Patient_16", "Patient_17", "Patient_18", "Patient_19", "Patient_20",
    "Patient_21", "Patient_22", "Patient_23", "Patient_24", "Patient_25",
    "Patient_26", "Patient_27", "Patient_28", "Patient_29", "Patient_30",
    "Patient_31", "Patient_32", "Patient_33", "Patient_34", "Patient_35",
    "Patient_36", "Patient_37", "Patient_38", "Patient_39", "Patient_40",
    "Patient_41", "Patient_42", "Patient_43", "Patient_44", "Patient_45",
    "Patient_46", "Patient_47", "Patient_48", "Patient_49", "Patient_50",
    "Patient_51", "Patient_52", "Patient_53", "Patient_54", "Patient_55",
    "Patient_56", "Patient_57", "Patient_58", "Patient_59", "Patient_60",
    "Patient_61", "Patient_62", "Patient_63", "Patient_64", "Patient_65",
    "Patient_66", "Patient_67", "Patient_68", "Patient_69", "Patient_70",
    "Patient_71", "Patient_72", "Patient_73", "Patient_74", "Patient_75",
    "Patient_76", "Patient_77", "Patient_78", "Patient_79", "Patient_80",
    "Patient_81", "Patient_82", "Patient_83", "Patient_84", "Patient_85",
    "Patient_86", "Patient_87", "Patient_88", "Patient_89", "Patient_90",
    "Patient_91", "Patient_92", "Patient_93", "Patient_94", "Patient_95",
    "Patient_96", "Patient_97", "Patient_98", "Patient_99", "Patient_100"]
sex = [
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male",
    "Female", "Male", "Female", "Male", "Female",
    "Male", "Female", "Male", "Female", "Male"]

pcv_results = [
    34, 45, 37, 56, 32, 41, 49, 39, 35, 52,
    44, 38, 29, 47, 36, 55, 33, 42, 46, 31,
    40, 53, 34, 48, 37, 57, 35, 43, 45, 30,
    39, 50, 33, 54, 36, 41, 47, 32, 38, 51,
    35, 44, 34, 56, 37, 40, 49, 31, 36, 52,
    43, 39, 30, 48, 35, 55, 34, 42, 46, 33,
    41, 53, 32, 47, 36, 57, 35, 44, 45, 31,
    40, 50, 34, 54, 37, 41, 48, 30, 39, 51,
    35, 43, 33, 56, 36, 40, 49, 32, 38, 52,
    44, 39, 31, 47, 35, 55, 34, 42, 46, 33
]
```

```python
#Initialize all counters to 0
anaemic_count = 0
normal_count = 0
polycythaemic_count = 0

male_count = 0
female_count = 0

male_anaemic = 0
male_normal = 0
male_polycythaemic = 0

female_anaemic = 0
female_normal = 0
female_polycythaemic = 0

print("============================================================")
print("PCV CLASSIFICATION REPORT FOR 100 PATIENTS")
print("============================================================")

for name, gender, pcv in zip(patient_ID, sex, pcv_results):
    classification = classify_pcv(gender, pcv)
    print(f"{name} | {gender} | {pcv}% | {classification}")
    
    if gender == "Male":
        male_count += 1
    elif gender == "Female":
        female_count += 1

    if classification == "Anaemic":
        anaemic_count += 1
        if gender == "Male":
            male_anaemic += 1
        elif gender == "Female":
            female_anaemic += 1
            
    elif classification == "Normal":
        normal_count += 1
        if gender == "Male":
            male_normal += 1
        elif gender == "Female":
            female_normal += 1
            
    else:
        polycythaemic_count += 1
        if gender == "Male":
            male_polycythaemic += 1
        elif gender == "Female":
            female_polycythaemic += 1
```
