RESULT INTERPRETATION
Less than 70 mg/dL → Hypoglycemia,
70–99 mg/dL → Normal,
100–125 mg/dL → Prediabetes,
126 mg/dL and above → Diabetes

```python
patient_name = input("Enter patient name:")
glucose_input = input("Enter glucose result (mg/dL):")
glucose_result = float(glucose_input)
```

```python
def interpret_result(glucose_result):
    if glucose_result < 70:
        return ("Hypoglycemia")
    elif 70 <= glucose_result <= 99:
        return "Normal"
    elif 100 <= glucose_result <= 125:
        return "Prediabetes"
    else:
        return "Diabetes"
interpret_result(glucose_result)

print("\n========================")
print("LABORATORY REPORT")
print("========================")
print(f"Patient Name: {patient_name}")
print("Test: Fasting Blood Glucose")
print(f"Result: {glucose_result} mg/dL")
print(f"Interpretation: {interpret_result(glucose_result)}")
print("========================")
```
